```python
def get_categories(df, lexicon):
    categories = []
    for text in df["body"]:
        category_dict = lexicon.analyze(text, normalize=True)
        categories.append(category_dict)
    df["categories"] = categories
    return df
```

```python
#get categories per label and normalize 
category_sum_vent = dict()
category_sum_ns = dict()
for col in mh_cat_df.columns:
    if col != "label":
        subset_vent = mh_cat_df[mh_cat_df["label"] == "Venting"]
        total_vent = subset_vent[col].sum()
        vent_size = len(mh_df[mh_df["label"] == "Venting"])
        category_sum_vent[col] = total_vent / vent_size
        subset_ns = mh_cat_df[mh_cat_df["label"] == "Need Support"]
        total_ns = subset_ns[col].sum()
        ns_size = len(mh_df[mh_df["label"] == "Need Support"])
        category_sum_ns[col] = total_ns / ns_size
```

```python
def get_top_categories(category_sum, number):
    sorted_cats = sorted(category_sum.items(), key=lambda x: x[1], reverse = True)
    top_dict = dict(sorted_cats[:number])
    return top_dict
```

```python
top_vent_cat = get_top_categories(category_sum_vent, 15)
top_ns_cat = get_top_categories(category_sum_ns, 15)
```

Number of stop words 

```python
def get_num_stop_words(df):
    nltk.download('stopwords')
    stop_words = set(stopwords.words('english'))
    stop_word_count = []
    for text in df["body"]:
        # Tokenize the text
        words = text.split()
        # Count the number of stopwords
        num_stopwords = len([word for word in words if word.lower() in stop_words])
        stop_word_count.append(num_stopwords)
    df["num_stopwords"] = stop_word_count
    return df
```
